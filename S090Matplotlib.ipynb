```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
y = x **2
print(x,y)

plt.plot(x,y, color='r', label = 'first') # var len args and keyword args
plt.plot(x,y**2, label = 'second') # var len args and keyword args
plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = x **2
print(x,y)

plt.plot(x,y,'go',x,y**2, 'r^', label = 'abcd')
plt.xlabel('time')
plt.ylabel('Price')
plt.title('Title')
plt.xticks(range(1,10,2))
plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = x **2
print(x,y)

plt.figure(figsize = (15,5))

plt.subplot(2,2, 1)
plt.plot(x,y, color='r') # var len args and keyword args

plt.subplot(2,2, 2)
plt.plot(x,y**2) # var len args and keyword args

plt.subplot(2,2, 3)
plt.plot(x,y**2) # var len args and keyword args

plt.subplot(2,2, 4)
plt.plot(x,y**2) # var len args and keyword args
plt.show()
```

```python
x = np.arange(1,11)
y = x **2
print(x,y)

plt.plot(x,y,'go',x,y**2, 'r^', label = 'abcd')
plt.text(10,20,'(10,20)')
plt.show()
```

```python
x = np.arange(1,11)
y = x **2
plt.stem(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,11)
y = x **2
y1 = x**3
plt.fill_between(x,y1, alpha = 0.4)
plt.fill_between(x,y, alpha = 0.4)
plt.axvline(4)
plt.axhline(60)
```
